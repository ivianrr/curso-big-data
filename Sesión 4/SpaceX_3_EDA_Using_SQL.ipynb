## Tarea 2: Queries con SQL

```python
def get_query(query):
  df  = pd.read_sql(query, con = engine )
  return df
```

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
q="""
SELECT DISTINCT LaunchSite
FROM SPACEXTBL
""" 
get_query(q)
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
get_query(
"""
SELECT FlightNumber, Date, BoosterVersion, Payload, Orbit, LaunchSite
FROM SPACEXTBL
WHERE LaunchSite LIKE 'CCS%'
LIMIT 5
""" 
)
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
get_query(
"""
SELECT SUM(PayloadMass) AS total_mass
FROM SPACEXTBL
WHERE Customers = 'NASA (CRS)'
""" 
)
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
get_query(
"""
SELECT AVG(PayloadMass) AS avg_mass
FROM SPACEXTBL
WHERE Serial LIKE 'B1%'
""" 
)
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
get_query(
"""
SELECT FlightNumber, MIN(Date), Payload, Orbit
FROM SPACEXTBL
WHERE LandingOutcome = 'True RTLS'
""" 
)
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
get_query(
"""
SELECT Serial,FlightNumber, Payload, Orbit
FROM SPACEXTBL
WHERE 
  PayloadMass>4000 AND PayloadMass<6000
  AND LandingOutcome='True ASDS'
""" 
)
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
get_query("""
SELECT
  CASE 
    WHEN Mission_Outcome=1 THEN 'Success'
    WHEN Mission_Outcome=0 THEN 'Failure'
    END AS Mission_Result,
  COUNT(FlightNumber) AS num
FROM SPACEXTBL
GROUP BY Mission_Outcome
""")
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
get_query(
"""
SELECT Serial,FlightNumber, Payload, Orbit, PayloadMass 
FROM SPACEXTBL
WHERE PayloadMass = (SELECT MAX(PayloadMass) FROM SPACEXTBL)
""" 
)
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
get_query("""
SELECT 
  Date,
  LandingOutcome,
  Serial,
  LaunchSite
FROM SPACEXTBL
WHERE EXTRACT(YEAR FROM Date)=2015 AND (LandingOutcome LIKE 'False%' OR LandingOutcome LIKE 'None%')
""")
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
get_query("""
SELECT 
  CASE WHEN Mission_Outcome=1 THEN 'Success' ELSE 'Failure' END AS Outcome,
  COUNT(*) AS num
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Mission_Outcome
ORDER BY 2 DESC
""")
```
